```python
# Import all required libraries
from __future__ import division # For python 2.*

import numpy as np
import matplotlib.pyplot as plt
from io import StringIO

from scipy.linalg import svd

np.random.seed(0)
%matplotlib inline
```

```python
# Making charts for milestone 1

data1 = np.genfromtxt('charts/milestone1_wallWalking/m1_map1_reward.txt', delimiter=', ')
data1=data1[1:-1]
#print(data.shape)

data2 = np.genfromtxt('charts/milestone1_wallWalking/m1_map2_reward.txt', delimiter=', ')
data2 = data2[1:-1]

data3 = np.genfromtxt('charts/milestone1_wallWalking/m1_map3_reward.txt', delimiter=', ')
data3 = data3[1:-1]

data = np.concatenate((data1,data2,data3), axis=0)

X = [i for i in range(1,data.shape[0] +1)]

start = 11
end = data.shape[0]+1
avg = 0
learning_rate = []

for i in range(start, end):
    avg = np.mean(data[i-10:i])
    learning_rate.append(avg)
X2 = [i for i in range(start,end)]
    
# print(len(X2))
# print(len(learning_rate))
# print(X2)

fig, ax = plt.subplots(1,1, figsize=(10,8))
ax.plot(X, data, 'r-', lw=3, label="Cumulative Rewards")
ax.plot(X2, learning_rate,'b-', lw=3, label="Learning Rate")

plt.title("Milestone 1: Results")
plt.xlabel("Episodes")
plt.ylabel("Cumulative Rewards")
ax.legend()
plt.show()
```
